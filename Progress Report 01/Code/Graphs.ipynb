```julia
using Plots
Plots.plotly();
Plots.default(size=(600,400));
Plots.default(label="");
Plots.default(ticks=:native);

using FFTW

using CSV
using DataFrames
df = CSV.read("experiment.csv", DataFrame);
```

```julia
fig1 = plot(df."Lines of Data", df."Compress Time (s)");
xlabel!("Lines of data");
ylabel!("Time (s)")
title!("Lines of data versus compression time")
display(fig1);
```

```julia
fig2 = plot(df."Lines of Data", df."Decompress Time (s)");
xlabel!("Lines of data");
ylabel!("Time (s)")
title!("Lines of data versus decompression time")
display(fig2);
```

```julia
fig3 = plot(df."Lines of Data", df."Compression Ratio");
xlabel!("Lines of data");
ylabel!("Compression Ratio")
title!("Lines of data versus compression ratio")
display(fig3);
```
